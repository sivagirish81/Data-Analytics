```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('D:\College\PES\Computer Science\Semester Wise\SEMESTER-3\Introduction to Data Science\Datasheets\PimaIndians.csv')
```

```python
#Dropping woman whose age is above 60 due to lack of physical exercise in their old age(ideal case)
df=df.drop(df[df.age>60].index)
df
```

```python
for index,row in df.glucose.iteritems():
    if (row<40):
        if (row>35):
            df.glucose.loc[index]=40
        else:
            df.glucose.loc[index]=np.median(df.glucose)
    elif (row>165):
        if (row<170):
            df.glucose.loc[index]=165
        else:
            df.glucose.loc[index]=np.median(df.glucose)
```

```python
for index,row in df.diastolic.iteritems():
    if (row<40):
        if (row>35):
            df.diastolic.loc[index]=40
        else:
            df.diastolic.loc[index]=np.median(df.diastolic)
    elif (row>100):
        if (row<105):
            df.diastolic.loc[index]=100
        else:
            df.diastolic.loc[index]=np.median(df.diastolic)
```

```python
for index,row in df.triceps.iteritems():
    if (row<10):
        if (row>5):
            df.triceps.loc[index]=10
        else:
            df.triceps.loc[index]=np.median(df.triceps)
    elif (row>40):
        if (row<45):
            df.triceps.loc[index]=40
        else:
            df.triceps.loc[index]=np.median(df.triceps)
```
