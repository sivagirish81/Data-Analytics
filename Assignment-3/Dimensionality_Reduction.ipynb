```python
data=data.drop(data[data.Age>60].index)
data
```

```python
for index,row in data.Glucose.iteritems():
    if (row<40):
        if (row>35):
            data.Glucose.loc[index]=40
        else:
            data.Glucose.loc[index]=np.median(data.Glucose)
    elif (row>165):
        if (row<170):
            data.Glucose.loc[index]=165
        else:
            data.Glucose.loc[index]=np.median(data.Glucose)
```

```python
for index,row in data.BloodPressure.iteritems():
    if (row<40):
        if (row>35):
            data.BloodPressure.loc[index]=40
        else:
            data.BloodPressure.loc[index]=np.median(data.BloodPressure)
    elif (row>100):
        if (row<105):
            data.BloodPressure.loc[index]=100
        else:
            data.BloodPressure.loc[index]=np.median(data.BloodPressure)
```

```python
sns.barplot(x='Outcome',y='Age',data=data)
#Most of the people who are around the age of 28 do not have diabetes but most of the people who are around 35 have diabetes
```

```python
for index,row in data.SkinThickness.iteritems():
    if (row<10):
        if (row>5):
            data.SkinThickness.loc[index]=10
        else:
            data.SkinThickness.loc[index]=np.median(data.SkinThickness)
    elif (row>40):
        if (row<45):
            data.SkinThickness.loc[index]=40
        else:
            data.SkinThickness.loc[index]=np.median(data.SkinThickness)
```

```python
sns.barplot(x='Outcome',y='SkinThickness',data=data)
```
